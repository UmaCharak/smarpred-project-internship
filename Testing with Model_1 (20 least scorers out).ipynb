## Import modules 

```python
import pandas as pd
import numpy as np

from sklearn.utils import shuffle
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import accuracy_score
#import joblib
```

## Load and shuffle training data

```python
df = pd.read_csv('reordered_I.csv', sep=",", header='infer')
df = df.drop(columns = ['AT_TA_percent'])
train = shuffle(df, random_state=0)
train = train.drop(train.iloc[:, -61:-1], axis=1) 
```

```python
train.head()
```

```python
X_train = train.iloc[:,1:-1]
y_train = train.iloc[:, -1]
```

## Load test data

```python
df1 = pd.read_csv('reordered_test1.csv', sep=",", header='infer')
df1 = df1.drop(columns = ['AT_TA_percent'])
test1 = shuffle(df1, random_state=0)
test1 = test1.drop(test1.iloc[:, -61:-1], axis=1)
#test1.head()
```

```python
df2 = pd.read_csv('reordered_test2.csv', sep=",")
df2 = df2.drop(columns = ['AT_TA_percent'])
test2 = shuffle(df2, random_state=0)
test2 = test2.drop(test2.iloc[:, -61:-1], axis=1)
```

```python
df3 = pd.read_csv('reordered_test3.csv', sep=",")
df3 = df3.drop(columns = ['AT_TA_percent'])
test3 = shuffle(df3, random_state=0)
test3 = test3.drop(test3.iloc[:, -61:-1], axis=1)
```

```python
df4 = pd.read_csv('reordered_test4.csv', sep=",")
df4 = df4.drop(columns = ['AT_TA_percent'])
test4 = shuffle(df4, random_state=0)
test4 = test4.drop(test4.iloc[:, -61:-1], axis=1)
```

## Creating test sets

```python
t1 = test1.iloc[:, 1:-1]
t2 = test2.iloc[:, 1:-1]
t3 = test3.iloc[:, 1:-1]
t4 = test4.iloc[:, 1:-1]
```

```python
y1 = test1.iloc[:, -1]
y2 = test2.iloc[:, -1]
y3 = test3.iloc[:, -1]
y4 = test4.iloc[:, -1]
```

## Optimize Parameter with Cross Validation GridSearchCV()

### LinearSVC

```python
optimal_params_LSVC = GridSearchCV(LinearSVC(random_state=0), param_grid={'C': [0.1, 1, 10, 100, 1000, 1100, 1500, 2000]}, cv=4, scoring='accuracy')
optimal_params_LSVC.fit(X_train,y_train)
```

```python
best_parameters_LSVC = optimal_params_LSVC.best_params_
print(best_parameters_LSVC)
print(optimal_params_LSVC.best_score_)
```

```python
optimal_params_LSVC.score(t1,y1)
```

```python
optimal_params_LSVC.score(t2,y2)
```

```python
optimal_params_LSVC.score(t3,y3)
```

```python
optimal_params_LSVC.score(t4,y4)
```

### Decision Tree

```python
tree_para = {'criterion':['gini', 'entropy'], 'max_depth':np.arange(2,16)}
optimal_params_DT = GridSearchCV(DecisionTreeClassifier(random_state=0), param_grid=tree_para, cv=4, scoring='accuracy')
optimal_params_DT.fit(X_train,y_train)
#optimal_params.cv_results_
```

```python
best_parameters_DT = optimal_params_DT.best_params_
print(best_parameters_DT)
print(optimal_params_DT.best_score_)
```

```python
optimal_params_DT.score(t1, y1)
```

```python
optimal_params_DT.score(t2,y2)
```

```python
optimal_params_DT.score(t3,y3)
```

```python
optimal_params_DT.score(t4,y4)
```

### Random Forest 

```python
param_grid_rf = { 
    'n_estimators': [100, 200, 300, 400, 700],
    'max_depth':np.arange(2,16)
}

optimal_params_RF = GridSearchCV(RandomForestClassifier(random_state=0), param_grid=param_grid_rf, cv=4, scoring='accuracy')

optimal_params_RF.fit(X_train,y_train)
```

```python
best_parameters_RF = optimal_params_RF.best_params_
print(best_parameters_RF)
print(optimal_params_RF.best_score_)
```

```python
optimal_params_RF.score(t1, y1)
```

```python
optimal_params_RF.score(t2,y2)
```

```python
optimal_params_RF.score(t3,y3)
```

```python
optimal_params_RF.score(t4,y4)
```

### MLP

```python
param_grid_mlp={'hidden_layer_sizes': range(2,21),
           'activation':['relu', 'identity'],
           'solver': ['lbfgs','sgd', 'adam'],
            'learning_rate': ['constant', 'adaptive']}
optimal_params_MLP = GridSearchCV(MLPClassifier(random_state=0), param_grid=param_grid_mlp, cv=4, scoring='accuracy')
optimal_params_MLP.fit(X_train,y_train)
```

```python
best_parameters_MLP = optimal_params_MLP.best_params_
print(best_parameters_MLP)
print(optimal_params_MLP.best_score_)
```

```python
optimal_params_MLP.score(t1, y1)
```

```python
optimal_params_MLP.score(t2,y2)
```

```python
optimal_params_MLP.score(t3,y3)
```

```python
optimal_params_MLP.score(t4,y4)
```
